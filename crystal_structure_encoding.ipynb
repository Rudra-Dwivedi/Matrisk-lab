```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("C:/Users/Rudradwivedi/Desktop/Matrisk AI/data/material_properties/materials_clean.csv")
df.head()
```

```python
# Example if not already numeric
encoder = LabelEncoder()
df['crystal_system_code'] = encoder.fit_transform(df['crystal_system'])
df['category_code'] = encoder.fit_transform(df['category'])
```

```python
features = [
    'n_elements', 'spacegroup_number', 'formation_energy_per_atom_ev',
    'energy_above_hull_ev', 'band_gap_ev', 'bulk_modulus_gpa',
    'shear_modulus_gpa', 'poisson_ratio', 'density_g_cm3',
    'nsites', 'volume_a3', 'melting_point_k',
    'crystal_code', 'crystal_system_code', 'category_code'
]

X = df[features].values
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=3)   # 3D embeddings
embeddings = pca.fit_transform(X_scaled)

print("Embeddings shape:", embeddings.shape)
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
labels = kmeans.fit_predict(embeddings)

df['material_cluster'] = labels
```

```python
embeddings_df = pd.DataFrame(embeddings, columns=['embed1','embed2','embed3'])
final_df = pd.concat([df, embeddings_df], axis=1)
final_df.to_csv("C:/Users/Rudradwivedi/Desktop/Matrisk AI/data/material_properties/material_embeddings.csv", index=False)
```
