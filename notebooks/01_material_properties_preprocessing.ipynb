```python
# 1. Import libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# 2. Load dataset
df = pd.read_csv(r"C:\Users\Rudradwivedi\Desktop\Matrisk AI\data\material_properties\DS1_material_properties_5500.csv")

# 3. Inspect
df.head()
df.info()
```

```python
df = df.drop_duplicates()
df = df.bfill()   # backward fill
df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]
```

```python
# Normalize column names: lowercase, replace spaces with underscores
df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]
```

```python
# Drop duplicates
df = df.drop_duplicates()

# Fill missing values (forward fill as example)
df = df.bfill()
```

```python
# Encode crystal system and category
df['crystal_system'] = df['crystal_system'].astype('category')
df['crystal_system_code'] = df['crystal_system'].cat.codes

df['category'] = df['category'].astype('category')
df['category_code'] = df['category'].cat.codes

# Ensure boolean flags are integers (0/1)
df['is_metal'] = df['is_metal'].astype(int)
df['is_stable'] = df['is_stable'].astype(int)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numeric_cols = [
    'formation_energy_per_atom_ev',
    'energy_above_hull_ev',
    'band_gap_ev',
    'bulk_modulus_gpa',
    'shear_modulus_gpa',
    'poisson_ratio',
    'density_g_cm3',
    'nsites',
    'volume_a3',
    'melting_point_k'
]

df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```

```python
df.to_csv("../data/material_properties/materials_clean.csv", index=False)
```
